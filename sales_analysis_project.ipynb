```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\ajeeth\\Downloads\\Python_Diwali_Sales_Analysis-main\\Python_Diwali_Sales_Analysis-main\\Diwali Sales Data.csv",encoding = "unicode_escape")
```

```python
df
```

```python
df[['Status','unnamed1']].isnull().value_counts()
```

```python
df = df.drop(['Status','unnamed1'],axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df[df['Amount'].isnull()]
```

```python
df = df.dropna(subset=['Amount'])
```

```python
df
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
values = df['Gender'].value_counts(ascending=False).values
```

```python
axes = sns.countplot(data=df,x='Gender')
axes.bar_label(container=axes.containers[0] ,labels=values)
```

```python
df
```

```python
sns.barplot(data=df,y='Amount',x='Gender',estimator=np.sum)
```

```python
df.groupby(['Gender'])['Amount'].sum().reset_index()
```

```python
axes=sns.countplot(data=df,x='Age Group',hue='Gender')
for i in axes.containers:
    axes.bar_label(i)
```

```python
sales_age = df.groupby(['Age Group'])['Amount'].sum().reset_index().sort_values(by='Amount',ascending=False)
```

```python
sales_age
```

```python
axes = sns.barplot(data=sales_age,y='Amount',x='Age Group')
```

```python
df
```

```python
top_state_orders = df.groupby(['State'])['Orders'].sum().reset_index().sort_values('Orders',ascending=False).head(10)
```

```python
top_state_orders
```

```python
fig=plt.figure(figsize=(18,4))
sns.barplot(data=top_state_orders,x='State',y='Orders')
```

```python
top_state_sales = df.groupby(['State'])['Amount'].sum().reset_index().sort_values('Amount',ascending=False).head(10)
```

```python
top_state_sales
```

```python
fig=plt.figure(figsize=(18,4))
sns.barplot(data=top_state_sales,x='State',y='Amount')
```
